```python
import numpy as np
for file in filelist[:55]:
    
    with open(file, 'rb') as f:
        
        scene = pickle.load(f)
        offense_tensors = []
        defense_tensors = []
        player_id = []

        for agent in scene.agents:
            
            single_agent_array = np.array([
                [x, y, vx, vy]
                for x, y, vx, vy in zip(agent.x, agent.y, agent.vx, agent.vy)
            ])  # shape: [T, 5]


            single_agent_tensor = torch.tensor(single_agent_array.T, dtype=torch.float32).unsqueeze(0)  # shape: [1, 5, T]


            if agent.teamid != scene.def_teamid:
                offense_tensors.append(single_agent_tensor)
            else:
                defense_tensors.append(single_agent_tensor)
                
                
        offense_tensor = torch.cat(offense_tensors, dim=0) if offense_tensors else None  # [6, 5, T]
        defense_tensor = torch.cat(defense_tensors, dim=0) if defense_tensors else None  # [5, 5, T]

        scene_tensor = torch.cat([offense_tensor, defense_tensor], dim=0)  # shape: [11, 5, T]
        
        T = scene_tensor.shape[-1]
        
        # 获取所有 agent 的 x 坐标: shape = [N, T]
        x_coords = scene_tensor[:, 0, :]

        # 创建布尔掩码：每一帧是否所有 x 都 < 53，返回 [T]
        valid_mask = (x_coords < 53).all(dim=0)
        
        # 找到第一个满足条件的时间点
        valid_indices = torch.nonzero(valid_mask, as_tuple=False)

        if len(valid_indices) > 0:
            first_valid_t = valid_indices[0].item()
            print(f"第一个所有x<53的时间是: {first_valid_t},总长{T}")
        else:
            print("没有任何时刻满足所有x < 53")
```

```python
def build_hex_indexer(r=1.5, rect_width=52, rect_height=52):
    dx = 3 / 2 * r
    dy = np.sqrt(3) * r

    def hexagon_vertices(x_center, y_center, r):
        angles = np.linspace(0, 2 * np.pi, 7)
        x = x_center + r * np.cos(angles)
        y = y_center + r * np.sin(angles)
        return x, y

    def get_hexagonal_grid(width, height, r):
        grid = []
        for row in range(height):
            for col in range(width):
                x = col * dx
                y = row * dy
                if col % 2 == 1:
                    y += dy / 2
                hex_x, hex_y = hexagon_vertices(x, y, r)
                grid.append((x, y, hex_x, hex_y))
        return grid

    # 计算网格大小
    cols = int(np.ceil(rect_width / dx))
    rows = int(np.ceil(rect_height / dy))
    hex_grid = get_hexagonal_grid(cols, rows, r)

    # 构建 hex_index_map
    hex_index_map = {}
    for index, (x_center, y_center, _, _) in enumerate(hex_grid):
        hex_index_map[(x_center, y_center)] = index

    # 返回查找函数
    def get_hex_index(x, y):
        min_dist = float('inf')
        nearest_center = None
        if x > 55:
            return 600
        for (hx, hy) in hex_index_map.keys():
            dist = np.sqrt((hx - x) ** 2 + (hy - y) ** 2)
            if dist < min_dist:
                min_dist = dist
                nearest_center = (hx, hy)
        return hex_index_map[nearest_center]

    return get_hex_index
```
